# Mean, Min and Max Indian surface temperature data from 1915 to 2017 

```python
#Data format:
#['YEAR' 'JAN' 'FEB' 'MAR' 'APR' 'MAY' 'JUN' 'JUL' 'AUG' 'SEP' 'OCT' 'NOV' 'DEC' 'ANNUAL' 'JAN-FEB' 'MAR-MAY' 'JUN-SEP' 'OCT-DEC']

mean_india_temp_file='Data/Mean_Temp_IMD_2017.csv'
max_india_temp_file='Data/Max_Temp_IMD_2017.csv'
min_india_temp_file='Data/Min_Temp_IMD_2017.csv'

with open(mean_india_temp_file, 'r') as f:
    reader=csv.reader(f)
    mean_india_temp=np.asarray(list(reader)[1:])

with open(max_india_temp_file, 'r') as f:
    reader=csv.reader(f)
    max_india_temp=np.asarray(list(reader)[1:])
    
with open(min_india_temp_file, 'r') as f:
    reader=csv.reader(f)
    min_india_temp=np.asarray(list(reader)[1:])

years_available=mean_india_temp[:, 0]    

mean_india_temp=mean_india_temp[:,13]
min_india_temp=min_india_temp[:, 13]
max_india_temp=max_india_temp[:, 13]
```

```python
fig, ax=plt.subplots(1, 1)

plt.plot(years_available, mean_india_temp, label='Mean surface temp')
plt.plot(years_available, min_india_temp, label='Min surface temp')
plt.plot(years_available, max_india_temp, label='Max surface temp')
x_labels=ax.get_xticks()
year=1900
for i in range(1, len(x_labels)):
    x_labels[i]=str(int(year+15))
    year+=17

ax.xaxis.set_major_locator(plt.MaxNLocator(7))
ax.yaxis.set_major_locator(plt.MaxNLocator(7))

ax.set_xticklabels(x_labels)

plt.ylabel('Temperature')
plt.xlabel('Years')
plt.legend(loc='upper center', bbox_to_anchor=(0.52, 1.15), ncol=3, fancybox=True, shadow=True)
```

```python
north_hemisphere_yearly_ice_data=north_hemisphere_ice_data['01-01-1979':'31-12-2016'].resample('12M').mean()[1:]
south_hemisphere_yearly_ice_data=south_hemisphere_ice_data['01-01-1979':'31-12-2016'].resample('12M').mean()[1:]

total_hemisphere=north_hemisphere_yearly_ice_data.loc[:, 'Extent'].add(south_hemisphere_yearly_ice_data.loc[:, 'Extent'])
plt.figure()
plt.plot(north_hemisphere_yearly_ice_data.Year, north_hemisphere_yearly_ice_data['Extent'], label='Northern Hemisphere', color='red')
plt.plot(south_hemisphere_yearly_ice_data.Year, south_hemisphere_yearly_ice_data['Extent'], label='Southern Hemisphere', color='blue')
plt.xlabel('Years')
plt.ylabel('Extent of ice coverage (in million sq km)')
plt.legend(loc='upper center', bbox_to_anchor=(0.52, 1.15), ncol=3, fancybox=True, shadow=True)

plt.figure()
total_hemisphere.plot(label='Total', color='green')
plt.xlabel('Years')
plt.ylabel('Extent of total ice coverage (in million sq km)')
```
