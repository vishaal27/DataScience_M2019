```python
from sklearn import datasets
iris=datasets.load_iris()
len(iris.data)
labels=iris.target
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3, random_state=0).fit(iris.data)
type(kmeans.labels_)


kmeans_labels=[]
for i in range(len(list(kmeans.labels_))):
    if(kmeans.labels_[i]==0):
        kmeans_labels.append('Iris-versicolor')
    elif(kmeans.labels_[i]==1):
        kmeans_labels.append('Iris-setosa')
    else:
        kmeans_labels.append('Iris-virginica')
```
